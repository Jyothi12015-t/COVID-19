```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
population=pd.read_excel('Indian States Population and Area.xlsx', engine='openpyxl')
```

```python
population
```

```python
covid=pd.read_csv("state_wise.csv")
```

```python
covid
```

```python
covid.columns
```

```python
covid = covid[['State', 'Confirmed', 'Recovered', 'Deaths', 'Active']]
```

```python
covid
```

```python
top5_Active=covid.sort_values(by="Active",ascending=False)
```

```python
sns.set(rc={'figure.figsize':(16,12)})
sns.barplot(x="State",y="Active",data=top5_Active[0:5],hue="State")
plt.show()
```

```python
top5_Deaths=covid.sort_values(by="Deaths",ascending=False)
```

```python
sns.set(rc={'figure.figsize':(16,12)})
sns.barplot(x="State",y="Deaths",data=top5_Deaths[0:5],hue="State")
plt.show()
```

```python
population=population.rename(columns={"Aadhaar assigned as of 2019": "POPULATION"})
```

```python
i = covid[covid.State  == 'Total'].index
covid=covid.drop(i)
```

```python
modified_data=pd.merge(covid, population, on='State')
```

```python
modified_data
```

```python
modified_data.sort_values(["Active","POPULATION"], ascending = (False, False))
```

```python
age=pd.read_csv("patients_data.csv")
```

```python
age
```

```python
age.columns
```

```python
age = age[['Age Bracket','Gender', 'Detected City',
       'Detected District', 'Detected State', 'Backup Notes']]
```

```python
modiage=age.rename(columns={"Detected District": "District"})
modiage=modiage.rename(columns={"Detected State": "State"})
```

```python
modiage
```

```python
modyage=modiage.dropna()
```

```python
modydata=pd.merge(modified_data, modyage, on='State')
```

```python
modydata 
```

```python
males=(modydata['Gender']=='M').sum()
females=(modydata['Gender']=='F').sum()
males
```

```python
a=[males,females]
```

```python
fig = plt.figure(figsize =(10, 7)) 
plt.rcParams['font.size'] = 20
plt.pie(a,labels=['Males','Females'],colors=['blue','red'],autopct='%1.1f%%')
plt.axis("equal")
plt.title("Gender Proportion")
```

```python
data_types_dict = {'Age Bracket': int} 
modydata = modydata.astype(data_types_dict) 
```
